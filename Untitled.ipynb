```python
def test_and_write_to_file(model, val_transform, subset='train', output_dir = "outputs/42", THRESHOLD=0.5):
    """
    writes and save to file iamges from each subswet
    """

    test_dataset = Dataset(
    root = '../deep_watershed/data/for_training/combined_aug_new',
    subset = subset,
    num_classes=1,
    transform=val_transform)


    test_loader = torch.utils.data.DataLoader(
        test_dataset,
        batch_size=1,
        shuffle=False,
        num_workers=0,
        drop_last=False)

    ious = []
    with torch.no_grad():
        for input, target, meta in tqdm(test_loader, total=len(test_dataset)):

            input = input.to(device=device, dtype=torch.float32)#.unsqueeze(0)
            target= target.to(device=device, dtype=torch.float32)#.unsqueeze(0)

            output = model(input)
            iou = iou_score(output, target)
            ious.append(iou)
            output_fmt = os.path.join(output_dir, subset)

            masks = output.squeeze(1).detach().cpu().numpy()
            for index, img_name in enumerate(meta['img_id']):

                output_fmt = os.path.join(output_fmt, img_name)
                cv2.imwrite(output_fmt, (masks[index]>THRESHOLD).astype(np.uint8)*255)
    ious = np.mean(ious)
    print("Mean IoU: {}".format(ious))
    return ious
```

```python
ious_train = test_and_write_to_file(model, val_transform, subset='train', output_dir = "outputs/combined_new_aug")
```

```python
ious_val = test_and_write_to_file(model, val_transform, subset='val', output_dir = "outputs/combined_new_aug")
```

```python
ious_test = test_and_write_to_file(model, val_transform, subset='test', output_dir = "outputs/combined_new_aug")
```
